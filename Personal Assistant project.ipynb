```python
##importing modules and assigning variables

##importing for table representation
import pandas as pd
import re                                ##python regular expression library
pd.set_option('display.max_colwidth',200)

##importing for speech recognization
import speech_recognition as sr
import pyttsx3
import webbrowser as wb

r=sr.Recognizer()          ##creating a variable for recognizing speech


##importing for selecting particular answer by calculating probability of the question and pairing it with the suitable answer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
##finding word frequency and creating our own question and finding its similarity
vectorizer = TfidfVectorizer(ngram_range= (1,3))
vec = vectorizer.fit_transform(convo['Question'])


##Asking our own question and finding its similarity
my_q = vectorizer.transform(['Will you destroy us ?'])
cs = cosine_similarity(my_q, vec)
rs = pd.Series(cs[0]).sort_values(ascending=False)
top5 = rs.iloc[0:5]

##printing top 5 related question and its similarity
print(top5)
convo.iloc[top5.index]['Question']
```

```python
##printing the answer of our own unique question

rsi= rs.index[0]
rsi
convo.iloc[rsi]['Answer']
```

```python
## creating necessary funtions for personal bot

## for searching
def search():
    print("Say what you want to search")
    engine1=pyttsx3.init()
    engine1.say("Say what you want to search")
    engine1.runAndWait()       
    with sr.Microphone() as source:
        search_audio = r.listen(source,0,3)
    
    search = r.recognize_google(search_audio)
    print(search)
    wb.open('https://www.google.co.in/search?q=%s'%search, new=2)

## for exiting chat
def exit():
    engine=pyttsx3.init()
    result=engine.say("Bye, hope you have a great day !")
    print("Bye, hope you have a great day !")
    engine.runAndWait()
    
## for random chat
def chat(answer):
    print(answer)
    engine=pyttsx3.init()
    voice = engine.say(answer)
    engine.runAndWait()
    print("_________________________________________________________________")
    print("Say something")
    main()

## for writing note
def write_note():
    try :
        print("Say what you want to note")
        engine2=pyttsx3.init()
        engine2.say("Say what you want to note")
        engine2.runAndWait()
        with sr.Microphone() as source:
            note_audio = r.listen(source,0,5)
        
        note = r.recognize_google(note_audio)
        ##Your notes will be saved in Notes.txt
        
        with open("Notes.txt", "+a") as f:  
            f.write(note)
        f1=open("Notes.txt", "+a")
        f1.write("\n")
        f1.close()
        engine1=pyttsx3.init()
        engine1.say("Done")
        engine1.runAndWait()
        main()
    except Exception as e:
        print(e)

## for reading note
def read_note():
    try:
        f2=open("Notes.txt", "r")
        read=f2.read()
        f2.close()
        engine1=pyttsx3.init()
        engine1.say(read)
        engine1.runAndWait()
        main()
    except Exception as e:
        print(e)
        
## for deleting note:
def delete_note():
    try:
        f3=open("Notes.txt", "w")
        f3.write("")
        f3.close()
        engine1=pyttsx3.init()
        engine1.say("done")
        engine1.runAndWait()
        main()
    except Exception as e:
        print(e)
        

## for opening email
def mail():
    print("Opening your e-mail")
    engine1=pyttsx3.init()
    engine1.say("Opening your e-mail. Please wait")
    engine1.runAndWait()       
    wb.open('https://mail.google.com/mail/u/0/?tab=rm0&ogbl#inbox', new=2)
        

## creating main funtion for personal bot

def main():
    with sr.Microphone() as source:
        
        audio = r.listen(source,0,3)     ##{0 and 3 are seconds to which it will wait for user to speak}
        print('Recognizing ... ')
    
    try:
        text = r.recognize_google(audio)
        print(text)
        my_que=vectorizer.transform([text])
        cs = cosine_similarity(my_que, vec)
        rs = pd.Series(cs[0]).sort_values(ascending = False)
        rsi = rs.index[0]
        answer=convo.iloc[rsi]['Answer']
        
        if text == "exit":
            exit()
            
        elif text == "search":
            search()
        
        elif text == "write note":
            write_note()
        
        elif text == "read note":
            read_note()
        
        elif text == "delete note":
            delete_note()
        
        elif text == "open mail":
            mail()
            
        else:
            chat(answer)
            
    except Exception as e:
        return e
    
```
